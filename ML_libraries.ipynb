```
import cv2
```

```
group_photo = plt.imread("/content/grup_photo.png",format='PNG')
```

```
group_photo.shape
4
```

```
plt.imshow(group_photo)
```

```
model=cv2.CascadeClassifier("/content/haarcascade_frontalface_alt.xml")
```

```
faces= model.detectMultiScale(group_photo,1.2)
```

```
faces.shape
```

```
faces
```

```
for face in faces:
  x,y,w,h=face
  group_photo1=cv2.rectangle(group_photo,(x,y),(x+w,y+h),(255,0,0),4)
```

```
plt.imshow(group_photo1)
```
